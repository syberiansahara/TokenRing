```python
n0 = '5'
datakek = pd.read_table(r'%s\%s.txt' %(theroot, n0), 
                            sep=' ',
                            dtype={
                             'formats': ('i4', 'i4', 'f2',
                                    'f4', 'f4', 'f4', 
                                         'f4', 'S4', 'f2')}, 
                           names=('nnodes', 'nframes', 'tgen', 
                                       'texec', 'tmean', 'tstd', 
                                       'trpt', 'load', 'loadrate'))

nn = ['10', '15', '25', '50', '100']
for ind, n in enumerate(nn):
#     data = np.loadtxt(r'%s\%s.txt' %(theroot, n), delimiter=',',  
#                       dtype={'names': ('nnodes', 'nframes', 'tgen', 
#                                        'texec', 'tmean', 'tstd', 
#                                        'trpt', 'load', 'loadrate'),
#                              'formats': ('i4', 'i4', 'f2',
#                                     'f4', 'f4', 'f4', 
#                                          'f4', 'S4', 'f2')})
    datakek = datakek.append(pd.read_table(r'%s\%s.txt' %(theroot, n), 
                            sep=' ',
                            dtype={
                             'formats': ('i4', 'i4', 'f2',
                                    'f4', 'f4', 'f4', 
                                         'f4', 'S4', 'f2')}, 
                           names=('nnodes', 'nframes', 'tgen', 
                                       'texec', 'tmean', 'tstd', 
                                       'trpt', 'load', 'loadrate'),))     

# print datakek, type(datakek)
```

```python
Z = datawow.as_matrix(columns=['nnodes'])
X = datawow.as_matrix(columns=['loadrate'])

Y = datawow.as_matrix(columns=['trpt'])
# Z = datawow.as_matrix(columns=['tmean'])

fig = plt.figure(figsize=[15,10])
ax = fig.gca(projection='3d')

# Make data.
X, Y = np.meshgrid(X, Y)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
# ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.view_init(elev=30, azim=140)  #  color = 'DarkMagenta',      
ax.dist=10

plt.show()
```
